```python
import ollama
```

```python
from ollama import chat
from pydantic import BaseModel

class InsightsModel(BaseModel):
    address : str
    abn : int
    website : str
    ICN : int
```

```python
unstructured_text = """
KPMG Australia is a leading professional services firm committed to delivering excellence across
a range of industries. With team of 250,000 employees worldwide and 17 locations across Australia.

We are registered business on the ICN Gateway (Business ID: 77569) and operate under ABN 
51 194 660 183. For inquires, contact us at +61 2 6248 1111
at kpmg.com/au/en/home.html
"""
```

```python
print(unstructured_text)
```

```python
prompt = f"""You are part of business records kepping system, extract key values from business listings:
        {unstructured_text}
        """
```

```python
print(prompt)
```

```python
res = chat(model='granite3.2:2b', messages=[
    {
        'role': 'user',
        'content': unstructured_text,
    }
], format= InsightsModel.model_json_schema())
```

```python
print(res ['message']['content'])
```

```python
import pandas as pd
import json

out = json.loads(res ['message']['content'])
pd.DataFrame(out, index=[0])
```
